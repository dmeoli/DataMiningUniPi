# Extract dataset and preprocess

```python
df = pd.read_csv('ds//new_customer_supermarket.csv', sep='\t')
df = df.loc[:, ['BasketID', 'BasketDate', 'CustomerID', 'ProdID', 'ProdDescr']]
df
```

```python
df = df.astype({'BasketDate': 'datetime64',
                'BasketID': 'object',
                'CustomerID': 'object'})
df.dtypes
```

```python
df.sort_values('BasketDate', inplace=True)
df['BasketDayOfYear'] = df['BasketDate'].dt.dayofyear
df = df.loc[:, ]
```

From the previous analisys we can state that any customer made two baskets on the same day, meaning that we can drop the attribute ***BasketID***

```python
id_descr = df[['ProdID', 'ProdDescr']]
id_descr = id_descr.drop_duplicates()
id_descr
```

```python
df2 = df[['CustomerID', 'ProdID', 'BasketDayOfYear']]
df2
```

```python
dataset = []
for cust, data in df2.groupby(df2['CustomerID']):
    seq = []
    for day, data2 in data.groupby(data['BasketDayOfYear']):
        itemset = np.sort(data2['ProdID'].values)
        element = (itemset, day)
        seq.append(element)
    dataset.append(seq)
```

# Basic Apriori
* minsup = 0.05
* No maxspan
* No maxgap
* No mingap

```python
run = False
if run:
    frequent_pattern = apriori_tc(dataset, minsup=0.05)
    res = pd.DataFrame(frequent_pattern, columns=['sequence', 'support'])
    res.to_csv('results_apriori_minsup_005')
else:
    res = pd.read_csv('results_apriori_minsup_005', sep=',')
    del res['Unnamed: 0']
res
```

# Apriori with time constraint
* minsup = 0.05
* maxspan = 30
* mingap = 0
* maxgap = 7

```python
if run:
    frequent_pattern = apriori_tc(dataset, minsup=0.05, maxspan=30, mingap=0, maxgap=7)
    res = pd.DataFrame(frequent_pattern, columns=['sequence', 'support'])
    res.to_csv('results_apriori_minsup_005_maxspan_30_mingap_0_maxgap_7')
else:
    res = pd.read_csv('results_apriori_minsup_005_maxspan_30_mingap_0_maxgap_7', sep=',')
    del res['Unnamed: 0']
res
```
